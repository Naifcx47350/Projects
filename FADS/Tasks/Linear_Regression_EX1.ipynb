## Preprocess and clean the dataframe

```python

import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# Load the CSV file into a Pandas DataFrame.
df = pd.read_csv('car_dataset.csv')
```

```python
#Clean the data
for column in df.columns:
    if column != 'name':  # Exclude the specified column
        if df[column].dtype == 'object':  # Check if the column contains string/object df
            label_encoder = LabelEncoder()
            df[column] = label_encoder.fit_transform(df[column])


df.head()
```

## Use the linear regression closed-form solution and calculate the coefficients of the linear regression equation (theta) and the predictions (Y)

```python
def calculate_theta(X, y):
    theta = np.linalg.inv(X.T @ X) @ (X.T @ y)
    return theta


theta = calculate_theta(X_train, y_train)

theta = theta.reshape(-1, 1)

theta

theta.shape

predections = X_test @ theta
```
